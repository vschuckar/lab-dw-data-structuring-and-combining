# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import numpy as np
import pandas as pd
import functions_data_cleaning_formatting as fn

data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')

data.head()
```

```python
def drop_column(df: pd.DataFrame):
    '''
    This function creates a copy of the original df and drops a specific column if it exists
    in a Pandas data frame.

    Inputs:
    df: pd.DataFrame

    Output:
    Copy of the original df without the specific column
    '''
   
    df2 = df.copy()
    if 'unnamed:_0' in df2.columns: 
        df2 = df2.drop('unnamed:_0', axis = 1) 
    return df2
```

```python
data = drop_column(data) # remove unnamed:_0
```

```python
data.shape
```

```python
data.isna().sum() # to check if there are missing values
```

```python
display(round(((data.duplicated().sum())/data.shape[0]*100),2)) # check for duplicates
```

```python
data.drop_duplicates() # drop them but keeping the first as default
```

```python
column_to_rename = {'employmentstatus': 'employment_status'}
data = fn.column_rename(data, column_to_rename)
```

```python
round_column = ['customer_lifetime_value', 'total_claim_amount']
data = fn.to_round(data, round_column)
```

```python
round_int = ['months_since_last_claim', 'number_of_open_complaints']
data = fn.to_int(data, round_int)
```

```python
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'], errors = 'coerce')
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
mkt = data.groupby(['sales_channel']).agg({'customer_lifetime_value':'sum'})
mkt['customer_lifetime_value'] = mkt['customer_lifetime_value'].round(2)
display(mkt)
```

```python
# This summary table shows us that the most sales comes from the Agent sales channel and the lowest ones through the Web channel.
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
result = data.groupby(['gender', 'education']).agg({'customer_lifetime_value':'mean'})
result['customer_lifetime_value'] = result['customer_lifetime_value'].round(2)
display(result)
```

```python
# This table shows that female customers with high school or below education are the ones with the highest customer lifetime value.
# For male customers, the highest would be with Master, but it is pretty close to high school or below as well. 
# For both female and male customers, the lowest customer life time value comes from who has a doctor level of education. 
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
r2 = data.groupby(['month', 'policy_type'])['number_of_open_complaints'].sum().reset_index()
r2.head()
```

```python
long_format = pd.melt(r2, id_vars=['policy_type', 'month'], value_vars = ['number_of_open_complaints'], var_name = 'metric', value_name = 'total_open_complaints')
display(long_format)
```
